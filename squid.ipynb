```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Potassium

#Current
def I_K(m_K, V_m):
    E_K = 12
    G_K = 36

    return (m_K ** 4) * G_K * (V_m - E_K)

#Gating Variable
def delta_m_K(V_m, m_K):
    def alpha_K(V_m):
        return (0.01 * (V_m + 10)) / (np.exp((V_m + 10)/10) - 1)
    def beta_K(V_m):
        return 0.125 * np.exp(V_m / 80)

    a = alpha_K(V_m)
    b = beta_K(V_m)
    return (1 - m_K) * a - m_K * b
```

```python
# Sodium

def I_Na(m_Na, h_Na, V_m):
    G_Na = 120
    E_Na = -115

    return h_Na * (m_Na ** 3) * G_Na * (V_m - E_Na)

def delta_m_Na(V_m, m_Na):
    def alpha_Na(V_m):
        return (0.1 * (V_m + 25)) / (np.exp((V_m + 25)/10) - 1)
    def beta_Na(V_m):
        return 4 * np.exp(V_m / 18)

    a = alpha_Na(V_m)
    b = beta_Na(V_m)
    return (1 - m_Na) * a - m_Na * b

def delta_h_Na(V_m, h_Na):
    def alpha_h(V_m):
        return 0.07 * np.exp(V_m / 20)
    def beta_h(V_m):
        return 1 / (np.exp((V_m + 30)/10) + 1)

    a = alpha_h(V_m)
    b = beta_h(V_m)
    return (1 - h_Na) * a - h_Na * b
```

```python
#Leak
def I_leak(V_m):
    G_leak = 0.3
    E_leak = -10.6
    return G_leak * (V_m - E_leak)
```

```python
#Voltage
def delta_V_m(I_Na, I_K, I_leak, I_in):
    C = 1
    I_sum = I_in - I_Na - I_K - I_leak
    return I_sum / C
```

```python
def run_sim(time_step, run_time, input_amount=50, in_start=0.4, in_end=0.6):
    input_current = 0
    
    # Set inital values
    V_m = 0
    m_K = 0.3
    m_Na = 0.1
    h_Na = 0.5
    t = 0

    # Storage
    V_m_array = []
    m_K_array = []
    m_Na_array = []
    h_Na_array = []
    I_Na_array = []
    I_K_array = []
    I_leak_array = []
    I_in_array = []
    time = []

    time_range = int(run_time / time_step)
    start_index = int(time_range * in_start)
    end_index = int(time_range * in_end)
    
    for i in range(time_range):
        if i == time_range // 4:
            print("25%")
        if i == time_range // 2:
            print("50%")
        if i == (time_range // 4) * 3:
            print("75%")

        if i == start_index:
            input_current = input_amount
        
        if i == end_index:
            input_current = 0

        # Time stuff
        time.append(t)
        t += time_step
        
        # Calculate deltas
        Na_current = I_Na(m_Na, h_Na, V_m)
        K_current = I_K(m_K, V_m)
        leak_current = I_leak(V_m)
        
        d_V_m = delta_V_m(Na_current, K_current, leak_current, input_current)
        d_m_K = delta_m_K(V_m, m_K)
        d_m_Na = delta_m_Na(V_m, m_Na)
        d_h_Na = delta_h_Na(V_m, h_Na)
    
        #Update
        V_m += d_V_m * time_step
        m_K += d_m_K * time_step
        m_Na += d_m_Na * time_step
        h_Na += d_h_Na * time_step
    
        #Store
        V_m_array.append(V_m)
        m_K_array.append(m_K)
        m_Na_array.append(m_Na)
        h_Na_array.append(h_Na)
        I_Na_array.append(Na_current)
        I_K_array.append(K_current)
        I_leak_array.append(leak_current)
        I_in_array.append(input_current)

    return [V_m_array, m_K_array, m_Na_array, h_Na_array, time, I_Na_array, I_K_array, I_leak_array, I_in_array]
```

```python
def show_data(res, start=0, end=-1):
    plt.plot(res[4][start:end], res[0][start:end], label='V_m')
    plt.legend(loc='best')
    plt.show()
    plt.plot(res[4][start:end],res[1][start:end], label='m_K')
    plt.plot(res[4][start:end],res[2][start:end], label='m_Na')
    plt.plot(res[4][start:end],res[3][start:end], label='h_Na')
    plt.legend(loc='best')
    plt.show()
    plt.plot(res[4][start:end],res[5][start:end], label='I_Na')
    plt.plot(res[4][start:end],res[6][start:end], label='I_K')
    plt.plot(res[4][start:end],res[7][start:end], label='I_leak')
    plt.plot(res[4][start:end],res[8][start:end], label='I_in')
    plt.legend(loc='best')
    plt.show()
```

```python
def run_sim_simple(time_step, run_time):
    input_current = 0
    
    # Set inital values
    V_m = -50
    t = 0

    # Storage
    V_m_array = []
    I_leak_array = []
    I_in_array = []
    time = []

    time_range = int(run_time / time_step)
    for i in range(time_range):
        if i == time_range // 4:
            print("25%")
        if i == time_range // 2:
            input_current = 50
            print("50%")
        if i == (time_range // 4) * 3:
            input_current = 0
            print("75%")

        # Time stuff
        time.append(t)
        t += time_step
        
        # Calculate deltas
        leak_current = I_leak(V_m)
        
        d_V_m = delta_V_m(0, 0, leak_current, input_current)
    
        #Update
        V_m += d_V_m * time_step
    
        #Store
        V_m_array.append(V_m)
        I_leak_array.append(leak_current)
        I_in_array.append(input_current)

    return [V_m_array, I_leak_array, I_in_array, time]
```

```python
res = run_sim(0.001, 50, -10, 0.2, 0.8)
```
