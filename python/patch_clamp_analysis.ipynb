```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import json
```

```python
sns.set_theme(style="whitegrid")
```

```python
with open("rust/final_results/cook_full_clamp.json", "r") as file:
    cook = np.array(json.load(file))
```

```python
voltage = cook[:, :, 0]
current = cook[:, :, 1] / 1000.0

plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)

plt.title("(B) Full Connectome")
plt.xlabel("Membrane Potential (mV)")
plt.ylabel("Current (A)")

for i in range(0, 280, 13):
    plt.plot(voltage[i, :], current[i, :])

plt.xlim(-150, 100)
#plt.ylim(y_min, y_max)

fig = plt.gcf()
fig.set_size_inches(4, 3)
plt.tight_layout()

plt.savefig('full_connectome.png', dpi=150)
plt.show()
```

```python
with open("rust/final_results/two_node_clamp.json", "r") as file:
    clamp = np.array(json.load(file))
```

```python
clamp.shape
```

```python
voltage = clamp[:, :, 0]
current = clamp[:, :, 1] / 1000.0

plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)

plt.title("(A) Two Neurons")
plt.xlabel("Membrane Potential (mV)")
plt.ylabel("Current (A)")

for i in range(1, 20, 4):
    plt.plot(voltage[i, :], current[i, :], label=f"{i}")

plt.xlim(-150, 100)
plt.ylim(-5, 30)

plt.legend(loc='best', title="No. Synapses")

fig = plt.gcf()
fig.set_size_inches(4, 3)
plt.tight_layout()

plt.savefig('img/two_node.png', dpi=150)
plt.show()
```
